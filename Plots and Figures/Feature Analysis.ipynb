```python
import sys
import time
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Raw_Results/IEEE30_IGratio.csv")
```

```python
data.head()
```

```python
ig = data["0"].to_numpy()
print(ig)
```

```python
features = range(len(ig))
features = np.asarray(features)
features = features + 1
print(features)
```

```python
SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 26

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
title_font = {'fontname':'Arial', 'size':'26', 'color':'black', 'weight':'bold',
              'verticalalignment':'bottom'} # Bottom vertical alignment for more space
```

```python
plt.figure(figsize=(16,10))
plt.scatter(features,ig, color = 'b')
plt.plot(features,ig, 'b-')



plt.ylabel('IG ratio')
plt.xlabel('Feature')
plt.title('Information Gain (IG) ratio of each feature of the IEEE 30-bus system', **title_font)
plt.grid()

plt.savefig('IEEE30_igratio.pdf')
plt.show()
```
